```python
#unique_values = cleaning_df['Hours to Complete'].unique()
#print(unique_values)

#cleaning_df['Hours to Complete'] = cleaning_df['Hours to Complete'].astype(float).astype(int)

#print(cleaning_df[['Course', 'Offered By', 'Rating', 'Summary', 'Level', 'Number of Reviews', 'Hours to Complete', 'Link to the Course']])

import re

def clean_and_convert_to_hours(value):
    if isinstance(value, (int, float)):
        return value

    # Initialize hours and minutes
    hours, minutes = 0, 0

    # Check for hours and minutes in the string
    hours_match = re.search(r'(\d+)\s*hours?', value)
    minutes_match = re.search(r'(\d+)\s*mins?', value)

    # Extract hours if present
    if hours_match:
        hours = int(hours_match.group(1))

    # Extract minutes if present and convert to hours
    if minutes_match:
        minutes = int(minutes_match.group(1)) / 60

    # Calculate total hours
    total_hours = hours + minutes

    # Handle cases where the value might be purely numerical in string format
    if total_hours == 0 and value.replace('.', '', 1).isdigit():
        return float(value)

    return total_hours

# Apply the cleaning and conversion function to the column
cleaning_df['Hours to Complete'] = cleaning_df['Hours to Complete'].apply(clean_and_convert_to_hours).astype(int)

# Display unique cleaned values
print(cleaning_df['Hours to Complete'].unique())



```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from googleapiclient.discovery import build

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# DataFrame from the latest version of the Coursera/Udemy.cvs
df = pd.read_csv('/content/coursera_udemy_2.csv')

# Function to preprocess text
def preprocess_text(text):
    tokens = word_tokenize(text.lower())  # Tokenization and convert to lowercase
    tokens = [token for token in tokens if token.isalnum()]  # Remove non-alphanumeric tokens
    tokens = [token for token in tokens if token not in stopwords.words('english')]  # Remove stopwords
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]  # Lemmatization
    return ' '.join(tokens)

# Function to search YouTube videos
def search_youtube_videos(api_key, preprocess_text, max_keywords=5):
    youtube = build('youtube', 'v3', developerKey=api_key)
    keywords = preprocess_text.split()
    search_query = ' '.join(keywords[:max_keywords])
    request = youtube.search().list(
        q=search_query,
        part='snippet',
        type='video',
        maxResults=10
    )
    response = request.execute()
    return response['items']

# Function to retrieve video details
def get_video_details(video_ids):
    youtube = build('youtube', 'v3', developerKey=api_key)
    request = youtube.videos().list(
        part='snippet,statistics',
        id=','.join(video_ids)
    )
    response = request.execute()
    return response['items']

# Function to present recommendations for YouTube videos
def present_recommendations(recommended_videos):
    for video in recommended_videos:
        print(f"YouTube Video: [{video['snippet']['title']}](https://www.youtube.com/watch?v={video['id']})\n")

# Input the sequence of text that was extracted from another model or write what you want to look for
pdf_text = "Describe one application of artificial intelligence in modern computing"
preprocessed_pdf_text = preprocess_text(pdf_text)

# Calculate TF-IDF vectors for course descriptions and PDF text
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(df['Summary'].values)
pdf_tfidf = tfidf_vectorizer.transform([preprocessed_pdf_text])

# Calculate cosine similarity between PDF text and course descriptions
similarities = cosine_similarity(pdf_tfidf, tfidf_matrix)

# indices of top similar courses
top_indices = similarities.argsort(axis=1)[:, -5:][:, ::-1]

# top recommended courses
top_courses = df.iloc[top_indices.flatten()]

# Print top recommended courses
styled_df = top_courses[['Course', 'Offered By', 'Summary', 'Rating', 'Number of Reviews', 'Level', 'Hours to Complete', 'Link to the Course']].style.set_properties(**{
    'background-color': 'lightgrey',
    'color': 'black',
    'border-color': 'white',
    'font-size': '10pt'
}).set_table_styles([{
    'selector': 'th',
    'props': [('background-color', 'darkgrey'), ('color', 'black'), ('font-size', '11pt')]
}])

# Define the YouTube API key
api_key = "AIzaSyAKXKLWqwllEDfqYU9SKMDEHIlHyJurwCQ"

# Call search_youtube_videos function to find relevant videos
videos = search_youtube_videos(api_key, preprocessed_pdf_text)

# Get video IDs
video_ids = [video['id']['videoId'] for video in videos]

# Get video details
video_details = get_video_details(video_ids)

# Recommend videos with most views
recommended_videos = sorted(video_details, key=lambda x: int(x['statistics']['viewCount']), reverse=True)
top_5_recommended_videos = recommended_videos[:5]

# Present recommendations for YouTube videos
present_recommendations(top_5_recommended_videos)

def save_recommendations_to_txt(recommended_videos, filepath):
    with open(filepath, 'w') as file:
        for video in recommended_videos:
            video_info = f"YouTube Video: [{video['snippet']['title']}](https://www.youtube.com/watch?v={video['id']})"
            print(video_info, file=file)

# Use the modified function to save the top 5 recommendations to a text file
save_recommendations_to_txt(top_5_recommended_videos, '/content/youtube_recommendations12.txt')
top_courses[['Course', 'Offered By', 'Summary', 'Rating', 'Number of Reviews', 'Level', 'Hours to Complete', 'Link to the Course']].to_excel('/content/courses_recommendations12.xlsx', index=False)

print(top_5_recommended_videos)

styled_df
```
