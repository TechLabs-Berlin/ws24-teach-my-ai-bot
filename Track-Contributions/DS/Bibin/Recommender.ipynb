```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.metrics.pairwise import linear_kernel
```

```python
features = ['Course Title', 'Skills', 'Modules', 'Subject', 'Hours to Complete', 'Level', 'What you will learn','Format','Number of Review']
```

```python
df.loc[:,'Combined Features'] =df[features].astype(str).agg(''.join,axis=1)
```

```python
df['Combined Features']
```

```python
df['Course Title'] = df['Course Title'].str.strip().str.lower()
```

```python
df.head(5)
```

```python
tfidf_vectorizer = TfidfVectorizer()
```

```python
tfidf_matrix = tfidf_vectorizer.fit_transform(df['Combined Features'])
```

```python
cosine_similarity = linear_kernel(tfidf_matrix,tfidf_matrix)
```

```python
import re

def recommend_course(course_title, cosine_sim=cosine_similarity):
    print("Input Title:", course_title)

    input_title = course_title.strip().lower()

    input_title = re.sub(r'[^\w\s]', '', input_title)

    input_title = input_title.encode('utf-8')

    indx = df[df['Course Title'].str.lower().str.encode('utf-8') == input_title].index
    print("Indices:", indx)

    if len(indx) == 0:
        print("Course title not found!")
        return None

    indx = indx[0]
    similarity_score = list(enumerate(cosine_sim[indx]))
    similarity_score = sorted(similarity_score, key=lambda x: x[1], reverse=True)
    similarity_score = similarity_score[1:11]
    print("Similarity Scores:", similarity_score)

    course_index = [i[0] for i in similarity_score]
    recommended = df.iloc[course_index][['Course Title', 'Link to the Course']]

    if len(recommended) == 0:
        print("No similar courses found!")
    print("Recommended Courses:", recommended)
    return recommended

recommended_courses = recommend_course("data security for web developers")
print(recommended_courses)



```
